```python
import json
```

```python
with open('builder_project_phase_unit_model_siting.json') as f:
    data = f.read() # reading the pre-constructed data
```

```python
print("Data type before reconstruction :", type(data)) # checking the container datatype of the pre-constructed data
```

```python
js = json.loads(data) # reconstructing the data from a string (a very long one by the looks of things) into a dictionary
```

```python
print("data type after reconstruction : ", type(js)) # Checking the data type container after re-construction
```

```python
js
```

```python
for key, value in js.items(): # Since there is only two initial keys (ProjectPhase and units) I extracted the value from the unit key.
    new_dict = value
```

```python
print(new_dict) # Here you can see the single value as a big list of dictionaries. Each item in the list represents a single unit.
```

```python
print(len(new_dict)) # It looks like there are 91 units.
```
